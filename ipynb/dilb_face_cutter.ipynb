```python
# 匯入必要的library
from scipy.io import loadmat
from skimage import io
import dlib, os, glob, re
from imutils import paths
import numpy as np
import cv2
import sklearn
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(r'C:/Users/User/Desktop/OpenCV/model/shape_predictor_68_face_landmarks.dat')

font = 1;    lt = 16            # 利用cv2.putText輸出1-68

img = cv2.imread('./makeup_subdata/makeup1.jpeg')     # cv2讀取影像
# img = cv2.imread('./image/lenaColor.png')     # cv2讀取影像


img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)  # 取灰度

height = img.shape[0]
posarray = []
rects = detector(img_gray, 0)             # 人臉數rects
for i in range(len(rects)):
    landmark = np.matrix([[p.x, p.y] for p in predictor(img,rects[i]).parts()])
    for idx, point in enumerate(landmark):
        
        pos = (point[0, 0], point[0, 1])   # 68點的座標
        posarray.append(pos)
posarray = np.array(posarray)
print(posarray)


# Display the masked image
cv2.imshow('Masked Image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
# Create position of each pixel
height, width, channels = img.shape
imgindx = [j for j in range(width)]
imgindy = [i for i in range(height)]
imgindx = np.array(imgindx)
imgindy = np.array(imgindy)

X = posarray[:16, 1].reshape(-1, 1)
y = -posarray[:16, 0]


# Fit polynomial regression model
degree = 4  # You can change the degree of the polynomial if needed
poly = PolynomialFeatures(degree)
X_poly = poly.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)

position_x_poly = poly.fit_transform(imgindx.reshape(-1,1))
position_y = poly_reg.predict(position_x_poly)

print(position_x_poly)

for i in imgindx:
    for j in imgindy: 
        if position_y[i]<-j:
            img[j,i,:]=0

for i in range(len(rects)):
    landmarks = np.matrix([[p.x, p.y] for p in predictor(img,rects[i]).parts()])
    for idx, point in enumerate(landmarks):
        
        pos = (point[0, 0], point[0, 1])   # 68點的座標
        print(idx,pos)

        cv2.circle(img, pos, 3, color=(0, 255, 0)) # 利用cv2.circle給每個特徵點畫一個圈，共68個
        cv2.putText(img, str(idx+1), pos, font, 0.8, (0, 0, 255), 1, lt)


# # Fit polynomial regression model
# # degree = 4  # You can change the degree of the polynomial if needed
# # poly = PolynomialFeatures(degree)
# # X_poly = poly.fit_transform(X)


cv2.imshow('Masked Image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
# # Fill the area below the polynomial curve on the mask
# cv2.fillPoly(mask, [pts], (255))

# Apply the mask to the image
```
